```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_br = pd.read_csv('BR_youtube_trending_data.csv')
df_ca = pd.read_csv('CA_youtube_trending_data.csv', on_bad_lines='skip', engine='python')
df_in = pd.read_csv('IN_youtube_trending_data.csv')
df_us = pd.read_csv('US_youtube_trending_data.csv')
```

```python
for df in [df_br, df_ca, df_in, df_us]:
    df.drop(columns=['description'], inplace=True)
```

```python
df_in.dropna(subset=['channelTitle'], inplace=True)
```

```python
df_br['region'] = 'BR'
df_ca['region'] = 'CA'
df_in['region'] = 'IN'
df_us['region'] = 'US'
```

```python
df_all = pd.concat([df_br, df_ca, df_in, df_us], ignore_index=True)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    if pd.isna(text):
        return 0
    return TextBlob(text).sentiment.polarity

df_all['title_sentiment'] = df_all['title'].apply(get_sentiment)
df_all['tags_sentiment'] = df_all['tags'].apply(get_sentiment)
```

```python
def label_sentiment(score):
    if score > 0.1:
        return 'Positive'
    elif score < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df_all['title_sentiment_label'] = df_all['title_sentiment'].apply(label_sentiment)
df_all['tags_sentiment_label'] = df_all['tags_sentiment'].apply(label_sentiment)
```

```python
import sqlite3

# database and connection
conn = sqlite3.connect(':memory:')

# Load DataFrame
df_all.to_sql('youtube', conn, index=False)

query = """
SELECT categoryId, AVG(view_count) AS avg_views
FROM youtube
GROUP BY categoryId
ORDER BY avg_views DESC
"""
result = pd.read_sql_query(query, conn)
```

```python
print(result)
```

```python
category_names = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    15: "Pets & Animals",
    17: "Sports",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "Howto & Style",
    27: "Education",
    28: "Science & Technology",
    29: "Nonprofits & Activism",

}

result['category_name'] = result['categoryId'].map(category_names)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=result.sort_values('avg_views', ascending=False),
            x='avg_views', y='category_name', palette='viridis')

plt.title('Average Views per Category')
plt.xlabel('Average Views')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
df_all['trending_date'] = pd.to_datetime(df_all['trending_date'], errors='coerce')
```

```python
trending_days = (
    df_all.groupby('video_id')['trending_date']
    .nunique()
    .reset_index()
    .rename(columns={'trending_date': 'days_trending'})
)
```

```python
df_all = df_all.merge(trending_days, on='video_id', how='left')
```

```python
top_videos = df_all[['video_id', 'title', 'days_trending']].drop_duplicates()
top_videos = top_videos.sort_values(by='days_trending', ascending=False).head(20)

plt.figure(figsize=(12, 8))
sns.barplot(y='title', x='days_trending', data=top_videos, palette='magma')
plt.title('Top 20 Longest Trending Videos')
plt.xlabel('Days Trending')
plt.ylabel('Video Title')
plt.tight_layout()
plt.show()
```

```python
avg_trending_by_category = (
    df_all.groupby('categoryId')['days_trending']
    .mean()
    .reset_index()
    .sort_values(by='days_trending', ascending=False)
)

avg_trending_by_category['category_name'] = avg_trending_by_category['categoryId'].map(category_names)

# Visualize
plt.figure(figsize=(12,6))
sns.barplot(data=avg_trending_by_category, x='days_trending', y='category_name', palette='coolwarm')
plt.title('Average Trending Duration by Category')
plt.xlabel('Average Days Trending')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
avg_trending_by_country = (
    df_all.groupby('region')['days_trending']
    .mean()
    .reset_index()
    .sort_values(by='days_trending', ascending=False)
)

plt.figure(figsize=(8,5))
sns.barplot(data=avg_trending_by_country, x='region', y='days_trending', palette='Spectral')
plt.title('Average Trending Duration by Country')
plt.xlabel('Country (Region)')
plt.ylabel('Average Days Trending')
plt.tight_layout()
plt.show()
```

```python
genre_views = (
    df_all.groupby('categoryId')['view_count']
    .sum()
    .reset_index()
    .sort_values(by='view_count', ascending=False)
)
genre_views['category_name'] = genre_views['categoryId'].map(category_names)

plt.figure(figsize=(12,6))
sns.barplot(data=genre_views, x='view_count', y='category_name', palette='Blues_d')
plt.title('Total Views by Genre (Category)')
plt.xlabel('Total Views')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
sentiment_by_category = (
    df_all.groupby('categoryId')[['title_sentiment', 'tags_sentiment']]
    .mean()
    .reset_index()
)
sentiment_by_category['category_name'] = sentiment_by_category['categoryId'].map(category_names)

plt.figure(figsize=(14,6))
sns.barplot(data=sentiment_by_category.sort_values('title_sentiment', ascending=False),
            x='title_sentiment', y='category_name', color='coral', label='Title Sentiment')
sns.barplot(data=sentiment_by_category.sort_values('tags_sentiment', ascending=False),
            x='tags_sentiment', y='category_name', color='skyblue', alpha=0.6, label='Tags Sentiment')

plt.title('Average Sentiment Scores by Genre')
plt.xlabel('Average Sentiment Polarity')
plt.ylabel('Genre')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sentiment_by_region = (
    df_all.groupby('region')[['title_sentiment', 'tags_sentiment']]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10,5))
sns.barplot(data=sentiment_by_region, x='region', y='title_sentiment', color='mediumseagreen', label='Title Sentiment')
sns.barplot(data=sentiment_by_region, x='region', y='tags_sentiment', color='slateblue', alpha=0.6, label='Tags Sentiment')

plt.title('Average Sentiment by Region')
plt.xlabel('Region')
plt.ylabel('Average Sentiment Polarity')
plt.legend()
plt.tight_layout()
plt.show()
```
